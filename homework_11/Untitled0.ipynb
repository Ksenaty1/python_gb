```python
import sympy as sp
```

```python
x = sp.Symbol("x")
expr = 5*x**2 + 10*x - 30
```

```python
# Определить корни (найти такие х при которых f(x) = 0)
sp.plot(expr)
expr_solve = sp.solve(expr)
expr_solve
```

```python
# Найти интервалы, на которых функция возрастает (f'(x) > 0)
sp.plot(sp.diff(expr))
expr_diff = sp.diff(expr)
sp.solve(expr_diff > 0)
```

```python
# Найти интервалы, на которых функция убывает (f'(x) < 0)
expr_diff = sp.diff(expr)
sp.solve(expr_diff < 0)
```

```python
# Построить график
sp.plot(expr)
```

```python
# Вычислить вершину (Такие х где f'(x) = 0)
sp.solve(expr_diff)
```

```python
# Определить промежутки, на котором f > 0 (решить неравенство f(x)>< 0)
sp.solve(expr > 0)
```

```python
# Определить промежутки, на котором f < 0
sp.solve(expr < 0)
```
